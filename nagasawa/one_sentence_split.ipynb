```python
# ファイル行数カウント
# with open('wiki_40b_test.txt', 'r') as f:
#     test_line_num = 0
#     for line in f:
#         test_line_num = test_line_num + 1
#     print("test_line_num:", test_line_num)
# f.close()

# with open('wiki_40b_val.txt', 'r') as f:
#     val_line_num = 0
#     for line in f:
#         val_line_num = val_line_num + 1
#     print("val_line_num:", val_line_num)
# f.close()

# with open('wiki_40b_train.txt', 'r') as f:
#     train_line_num = 0
#     for line in f:
#         train_line_num = train_line_num + 1
#     print("train_line_num:", train_line_num)
# f.close()
test_line_num = 103049
val_line_num = 103184
train_line_num = 1867351
```

```python
count = 0
with open("wiki_40b_test.txt", 'r') as f_read:
    with open("wiki_40b_test_one_stc.txt", 'w') as f_write:
        for line in f_read:
            count += 1
            stc_list = re.split("(?<=。)", line)
            stc_list.pop()
            for stc in stc_list:
                f_write.write("%s\n" % stc)
            progress = int(count / test_line_num * 100)
            if(count % 10000 == 0 or count == test_line_num):
                print("test:{} %".format(progress))
f_read.close()
f_write.close()
```

```python
count = 0
with open("wiki_40b_val.txt", 'r') as f_read:
    with open("wiki_40b_val_one_stc.txt", 'w') as f_write:
        for line in f_read:
            count += 1
            stc_list = re.split("(?<=。)", line)
            stc_list.pop()
            for stc in stc_list:
                f_write.write("%s\n" % stc)
            progress = int(count / val_line_num * 100)
            if(count % 10000 == 0 or count == val_line_num):
                print("val:{} %".format(progress))
f_read.close()
f_write.close()
```

```python
count = 0
with open("wiki_40b_train.txt", 'r') as f_read:
    with open("wiki_40b_train_one_stc.txt", 'w') as f_write:
        for line in f_read:
            count += 1
            stc_list = re.split("(?<=。)", line)
            stc_list.pop()
            for stc in stc_list:
                f_write.write("%s\n" % stc)
            progress = int(count / train_line_num * 100)
            if(count % 10000 == 0 or count == train_line_num):
                print("train:{} %".format(progress))
f_read.close()
f_write.close()
```
